```python
# export

import uproot3
import pandas as pd
import fnmatch
import itertools


def read_root_file(filename, branches=None):

    output_dict = {}

    with uproot3.open(filename) as f:
        # for now only read detector 1
        tree_name = "rqDir/zip1"
        selected_branches = [k.decode("utf-8") for k in f[tree_name].keys()]
        if branches is not None:
            selected_branches = list(
                itertools.chain.from_iterable(
                    fnmatch.filter(selected_branches, pattern) for pattern in branches
                )
            )
        df = next(
            uproot3.pandas.iterate(
                [filename], tree_name, selected_branches, entrysteps=float("inf")
            )
        )
        df = df.assign(
            DetNum=int(tree_name[9:]),
            EventNumber=f["rqDir/eventTree"]["EventNumber"].array(),
            SeriesNumber=f["rqDir/eventTree"]["SeriesNumber"].array(),
        )
    return df
```

```python
series_number = "25220213_171932"
filename = f"/cvmfs/data/CDMS/RQanalysis_testing/RQroot/OFResults_{series_number}.root"
branches = ["*IntegralHead*"]
```

```python
df = read_root_file(filename, branches=branches).head()
```

```python
assert len(df.columns) == 15
assert df.DetNum[0] == 1
assert df.SeriesNumber[0] == int(series_number.replace("_", ""))
```
